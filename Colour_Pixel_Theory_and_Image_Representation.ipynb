```python
# Loading the path of the image

image_path = '/content/drive/MyDrive/duck.jpg'

# Import an image
image = Image.open(image_path)

# Extracting pixel map:
pixel_matrix = image.load()
```

**Conversion of RGB to GrayScale**

The grayscale image is obtained by finding the average of the three channels of an RGB image.
   --> The Mathematical Formula to convert an RGB image to grayscale is as shown below:
                **( R + G + B) / 3**

```python
for i in range(width):  # x coordinate
  for j in range(height):  # y coordinate

    # getting the RGB pixel values.
    #  how much R is in it, how much G and how much B
    r, g, b = image.getpixel((i, j))

    # Apply formula of grayscale:
    grayscale = (r+g+b)/ 3    #grayscale normal average

    #setting the pixel value.
    pixel_matrix[i, j] = (int(grayscale), int(grayscale), int(grayscale))
```

```python
display(image)
```

This common formula for obtaining a grayscale is theoretically correct, but an improved formula has been ibtained by researchers from the weighted average of the pixel intensities of red, blue and green colours which can be shown as:



##                     ..................... (0.299R + 0.587G + 0.114B)  .................

```python
for i in range(width):  # x coordinate
  for j in range(height):  # y coordinate

    # getting the RGB pixel values.
    #  how much R is in it, how much G and how much B
    r, g, b = image.getpixel((i, j))

    # Apply formula of grayscale:
    grayscale = (0.299*r + 0.587*g + 0.114*b)   #grayscale normal average

    #setting the pixel value.
    pixel_matrix[i, j] = (int(grayscale), int(grayscale), int(grayscale))
```

```python
# Saving the final output as "grayscale.png"
image.save("grayscale.png", format = "png")
```
